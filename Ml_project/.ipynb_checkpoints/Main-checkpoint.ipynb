```python
import cv2
import csv
import glob
import pandas as pd

# Function to generate header
def generate_header():
    header = ["label"]
    for i in range(0, 784):
        header.append("pixel" + str(i))
    return header

# Function to convert image to binary values
def image_to_binary(img):
    binary_data = []
    rows, cols = img.shape
    for i in range(rows):
        for j in range(cols):
            k = img[i, j]
            if k > 100:
                k = 1
            else:
                k = 0
            binary_data.append(k)
    return binary_data

# Write header to CSV file
with open('dataset1.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(generate_header())

# Iterate over images and generate dataset
for label in range(10):
    dirList = glob.glob("captured_images/" + str(label) + "/*.png")

    for img_path in dirList:
        im = cv2.imread(img_path)
        im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
        im_gray = cv2.GaussianBlur(im_gray, (15, 15), 0)
        roi = cv2.resize(im_gray, (28, 28), interpolation=cv2.INTER_AREA)

        # Divide the ROI into 390 equal parts
        rows, cols = roi.shape
        step_size = rows // 390

        for i in range(0, rows, step_size):
            for j in range(0, cols, step_size):
                # Extract the region of interest
                roi_chunk = roi[i:i+step_size, j:j+step_size]

                data = [label]
                binary_data = image_to_binary(roi_chunk)
                data.extend(binary_data)

                # Append data to CSV file
                with open('dataset1.csv', 'a', newline='') as f:
                    writer = csv.writer(f)
                    writer.writerow(data)

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('dataset1.csv')

# Display the first few rows of the DataFrame
print(df.head())
```
